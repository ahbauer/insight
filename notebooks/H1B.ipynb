```python
timezp = 734500

def lnlike(params, t_e, economy, t_h, h1b, fit_line):
    [lag, m, b] = params
    if not fit_line:
        m=0
        b=0
    econ_shifted = np.interp(t_h, t_e+lag, economy+(m*t_e + b))
    chi2 = np.sum((econ_shifted-h1b)**2)
    return -0.5*chi2

def lnprob(params, t_e, economy, t_h, h1b, fit_line):
    lp = 0. #lnprior_v(theta)
    [lag, m, b] = params
    if lag < -400.0 or lag > 400.0 or m < 0.:
        return -np.inf
    return lp + lnlike(params, t_e, economy, t_h, h1b, fit_line)

def find_lag(input_series, fit_line=True): # input by year is good
    guess = [200.0, 0.0, 0.0] # time lag
    # nll = lambda *args: -lnlike(*args)
    # #results = optimize.fmin(nll, guess, args=(x_sorted, y_sorted, np.array(means)-500, h1b_binned), xtol=5., ftol=0.001, full_output=True)
    # results_ml = scipy.optimize.minimize(nll, guess, args=(x_sorted, y_sorted, np.array(means), h1b_binned), tol=0.01, bounds=[(-1000.0, 1000.0)])
    # print results_ml.x[0]

    ndim, nwalkers = len(guess), 100
    pos = [1e-4*np.random.randn(ndim) for i in range(nwalkers)]
    myargs =(np.array([d.toordinal()-timezp for d in input_series.index]), input_series.values, np.array([d.toordinal()-timezp for d in year_series_norm.index]), year_series_norm.values, fit_line)
    sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=myargs)
    pos, prob, state = sampler.run_mcmc(pos, 500)
    nsteps = 500
    print 'Burned in 500 steps, resetting and running {0} more.'.format(nsteps)
    sampler.reset()
    sampler.run_mcmc(pos, nsteps)
    print 'Finished running the sampler'
    print("Mean acceptance fraction: {0:.3f}".format(np.mean(sampler.acceptance_fraction)))
    # flatten the walkers
    samples = sampler.chain[:, :, :].reshape((-1, ndim))
    lags = samples[:,0]
    ms = samples[:,1]
    bs = samples[:,2]
    return lags, ms, bs
```

```python
# h1b_model = linear_regression(shifted_vals)

def make_x(indices,method='bfill'):
    month_array = indices.month
    ilow = month_array<=2
    ipeak1 = month_array==3
    ipeak2 = month_array==6
    ihigh = (month_array==4) | (month_array==5) | (month_array>6)
    month_array[ilow] = 1
    month_array[ipeak1] = 4
    month_array[ipeak2] = 2
    month_array[ihigh] = 1
    time_ordinal = [d.toordinal()-timezp for d in indices]
    time_ordinal = (time_ordinal-time_mean)/time_std
    in1 = sents_model.reindex(indices, method=method).values
    in2 = sp500_model.reindex(indices, method=method).values
    in3 = empl_model.reindex(indices, method=method).values
    in4 = trend_model.reindex(indices, method=method).values
    in5 = in1+1*time_ordinal
    in6 = in2+1*time_ordinal
    in7 = in3+1*time_ordinal
    in8 = in4+1*time_ordinal
    in9 = month_array
    in10 = time_ordinal
    return np.array([in1,in2,in3,in4,in5,in6,in7,in8,in9,in10]).T


# for testing:
indices = indices_tot #indices_train

time_ordinal = [d.toordinal()-timezp for d in indices]
time_mean = np.mean(time_ordinal)
time_std = np.std(time_ordinal)
time_ordinal = (time_ordinal-time_mean)/time_std
# for the real model:
#indices = sentiments_month['2004-01-01':'2014-10-01'].index

y_train = np.array(month_series_norm.reindex(indices, method='bfill').values)
#print y_data


feature_names = ['sentiment','SP500','unemployment','trends','sent+1','SP500+1','unempl+1','trends+1','month array']

#x_train = np.array([in1,in2,in3,in4,in5,in6,in7,in8,in9]).T

x_train = make_x(indices,method='bfill')

#print x_data
#regr = linear_model.Ridge()
# ds = []
# cs = []
# for depth in np.arange(2,10):
#     for iter in range(10):
#         #depth = 5
#         #regr = DecisionTreeRegressor(random_state=0,max_depth=depth)
#         regr = RandomForestRegressor(n_estimators=10, max_depth=depth)
#         crossval = cross_val_score(regr, x_data, y_data, cv=KFold(n=len(y_data), n_folds=10, shuffle=True))
#         ds.append(depth)
#         cs.append(np.mean(crossval))
    
#plt.plot(ds,cs,'o')

# depth=4
# n_trees = len(indices)-1
# regr = RandomForestRegressor(n_estimators=n_trees, min_samples_leaf=2,max_depth=depth) #min_samples_leaf=2,max_depth=depth)
# crossval = cross_val_score(regr, x_train, y_train, cv=KFold(n=len(y_train), n_folds=10, shuffle=False))
# print crossval
# print np.mean(crossval)

depth=3
n_trees=50
regr = GradientBoostingRegressor(n_estimators=n_trees,loss='ls',max_depth=depth,subsample=0.8)
crossval = cross_val_score(regr, x_train, y_train, cv=KFold(n=len(y_train), n_folds=10, shuffle=False))
print crossval
print np.mean(crossval)

# train_sizes, train_scores, test_scores = learning_curve(regr, x_data, y_data, cv=10)
# plt.plot(train_sizes, np.mean(train_scores,axis=1), label='Train Scores')
# plt.plot(train_sizes, np.mean(test_scores,axis=1), label='Test Scores')
# plt.legend()

# prange=range(2,15)
# train_scores, test_scores = validation_curve(regr, x_data, y_data, "max_depth", prange, cv=10) #, cv=10, scoring="accuracy", n_jobs=1)
# train_scores_mean = np.mean(train_scores, axis=1)
# train_scores_std = np.std(train_scores, axis=1)
# test_scores_mean = np.mean(test_scores, axis=1)
# test_scores_std = np.std(test_scores, axis=1)
# plt.plot(prange, train_scores_mean, label="Training score", color="r")
# plt.fill_between(prange, train_scores_mean - train_scores_std,
#                  train_scores_mean + train_scores_std, alpha=0.2, color="r")
# plt.plot(prange, test_scores_mean, label="Cross-validation score",
#              color="g")
# plt.fill_between(prange, test_scores_mean - test_scores_std,
#                  test_scores_mean + test_scores_std, alpha=0.2, color="g")
# plt.legend(loc="best")

#regr = RandomForestRegressor(n_estimators=n_trees, min_samples_leaf=2, max_depth=depth, oob_score=True)
#regr = GradientBoostingRegressor(n_estimators=n_trees,loss= 'huber')
regr.fit(x_train, y_train) #, sample_weight=month_array)
print feature_names
importances['computers'] = regr.feature_importances_
#print regr.oob_improvement_
#print regr.train_score_

#print regr.get_params()
joblib.dump(regr, '/Users/bauer/insight/teleborder/data/my_model.pkl', compress=9)
joblib.dump((sents_model.sort_index(), sp500_model.sort_index(), empl_model.sort_index(), trend_model.sort_index()), '/Users/bauer/insight/teleborder/data/xdata.pkl', compress=9)
#joblib.dump(sp500_model, '/Users/bauer/insight/teleborder/data/x2.pkl', compress=9)
#joblib.dump(empl_model, '/Users/bauer/insight/teleborder/data/x4.pkl', compress=9)
joblib.dump((norm_mean,norm_std,time_mean,time_std,np.median(lags_sentiment),np.median(lags),np.median(lags_empl),np.median(lags_trend)), '/Users/bauer/insight/teleborder/data/norms.pkl', compress=9)
#plt.plot(y_data, regr.predict(x_data), 'o', color='blue', linewidth=3)

fig = plt.figure(figsize=(10,8))
pred_train_series = pandas.Series(regr.predict(x_train),index=indices)
month_series_norm[indices].plot(label='H1B')
ax0=pred_train_series.plot(label='Prediction',color='red')
plt.xlabel('Date',fontsize='16')
plt.ylabel('# H1Bs', fontsize='16')
plt.title('Number of H1B Applications',fontsize='16')
plt.legend()
plt.setp(plt.gca().get_legend().get_texts(), fontsize='14')


```
